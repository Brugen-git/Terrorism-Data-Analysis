Exploratory Data Analysis - Terrorism

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
import warnings
warnings.filterwarnings("ignore")
data = pd.read_csv("D:/abcd/globalterrorismdb_0718dist.csv",encoding='latin1')
data.head()
```

```python
data.columns.values
```

```python
data.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude',
    'longitude':'longitude','summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound'},inplace=True)
```

```python
data = data[['Year','Month','day','Country','State','Region','City','latitude','longitude',"Attacktype",'kill',
               'Wound','target1','summary','Group','Targettype','Weapon','motive']]
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data['Wound'] = data['Wound'].fillna(0)
data['kill'] = data['kill'].fillna(0)
data['Casualities'] = data['kill'] + data['Wound']
data.info()
```

```python
data.describe()
```

```python
year = data['Year'].unique()
years_count = data['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (20,20))
sns.barplot(x = year,
           y = years_count,
           palette = "crest")
plt.xticks(rotation = 90,fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Attacking Year',fontsize=20)
plt.ylabel('Number of Attacks Each Year',fontsize=20)
plt.title('Attacks In Years',fontsize=30)
plt.show()
```

```python
pd.crosstab(data.Year, data.Region).plot(kind='area',stacked=False,figsize=(20,20))
plt.title('Terrorist Activities By Region In Each Year',fontsize=30)
plt.ylabel('Number of Attacks',fontsize=30)
plt.xlabel("Year",fontsize=30)
plt.xticks(rotation = 90,fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
attack = data.Country.value_counts()[:10]
```

```python
data.Group.value_counts()[1:10]
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(data['Country'].value_counts()[:10].index,data['Country'].value_counts()[:10].values,palette='Wistia_r')
plt.title('Top Countries Affected',fontsize=30)
plt.xlabel('Countries',fontsize=30)
plt.ylabel('Count',fontsize=30)
plt.xticks(rotation = 90,fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
df = data[['Year','kill']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(20,20))
df.plot(kind='bar',alpha=0.9,ax=ax4,color="purple")
plt.xticks(rotation = 90,fontsize=15)
plt.yticks(fontsize=15)
plt.title("People Died Due To Attack",fontsize=30)
plt.ylabel("Number of killed peope",fontsize=30)
plt.xlabel('Year',fontsize=30)
top_side = ax4.spines["top"]
top_side.set_visible(True)
right_side = ax4.spines["right"]
right_side.set_visible(False)
```

```python
data['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,20),color='red')
plt.xlabel("City",fontsize=20)
plt.ylabel("Number of attack",fontsize=20)
plt.xticks(rotation = 90,fontsize=15)
plt.yticks(fontsize=15)
plt.title("Top 10 most effected city",fontsize=30)
plt.show()
```

```python
data['Attacktype'].value_counts().plot(kind='bar',figsize=(20,20),color='grey')
plt.xticks(rotation = 90,fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Attacktype",fontsize=20)
plt.ylabel("Number of attack",fontsize=20)
plt.title("Name of Attack Type",fontsize=30)
plt.show()
```

```python
data[['Attacktype','kill']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(20,20),color=['orangered'])
plt.xticks(rotation = 90,fontsize=15)
plt.yticks(fontsize=15)
plt.title("Number of killed ",fontsize=30)
plt.ylabel('Number of people',fontsize=25)
plt.xlabel('Attack type',fontsize=20)
plt.show()
```

```python
data[['Attacktype','Wound']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(20,20),color=['crimson'])
plt.xticks(rotation = 90,fontsize=15)
plt.yticks(rotation = 90,fontsize=15)
plt.title("Number of wounded  ",fontsize=30)
plt.ylabel('Number of people',fontsize=20)
plt.xlabel('Attack type',fontsize=20)
plt.show()
```

```python
plt.subplots(figsize=(20,20))
sns.countplot(data["Targettype"],order=data['Targettype'].value_counts().index,palette="gist_heat",edgecolor=sns.color_palette("pastel"));
plt.xticks(rotation = 90,fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Attack type",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.title("Attack per year",fontsize=30)
plt.show()
```

```python
data['Group'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',color='cyan',figsize=(20,20))
plt.title("Top 10 terrorist group attack",fontsize=30)
plt.xlabel("Terrorist group name",fontsize=20)
plt.ylabel("Attack number",fontsize=20)
plt.xticks(rotation = 90,fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
data[['Group','kill']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('kill',ascending=False).head(10).plot(kind='bar',color='red',figsize=(20,10))
plt.title("Top 10 terrorist group attack",fontsize=30)
plt.xlabel("Terrorist group name",fontsize=20)
plt.ylabel("No of killed people",fontsize=20)
plt.xticks(rotation = 90,fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
